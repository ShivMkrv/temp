````python
# Providing Docstrings for functions
def least_diff(a, b, c):
  """
    Return the smallest difference between three values

    >>> least_diff(1,5,-5)
    4
  """
  d1 = abs(a - b)
  d2 = abs(b - c)
  d3 = abs(a - c)

  return min(d1, d2, d3)

print(least_diff(43, 23, 57))
help(least_diff)
````

````python
# Default values
def least_diff(a, b, c=10):
  d1 = abs(a - b)
  d2 = abs(b - c)
  d3 = abs(a - c)

  return min(d1, d2, d3)

print(least_diff(43, 23))
````

However, pandas has its own accessor operators, loc and iloc. For more advanced operations, these are the ones you're supposed to be using.

Both loc and iloc are row-first, column-second. This is the opposite of what we do in native Python, which is column-first, row-second.

iloc - index based selection

loc - label based selection

iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.

## Numpy

Numpy is used to create n-dim arrays. E.g. - 1D array, 2D array, 3D arrays etc.

How is Numpy different from Lists?
- Numpy is very fast. Because it uses Fixed types. 
- Numpy just contains the size, while the list contains size, reference count, object type and object value.
- Thus it is faster to read data from numpy.
- We also don't need to do type checking as all elements are of the same type.
- Numpy uses contiguous memory.
  - It helps in SIMD (Single Instruction Multiple Data) vector processing. e.g. If we have to do multiple additions on the list, then we can use SIMD vector to perform more than 1 additions at a time.
  - Effective Cache Utilization (shorter memory lookups).

```python
# lists
a = [1,2,3]
b = [4,5,6]

a * b    # Error
```

```python
# numpy
a = np.array([1,2,3])
b = np.array([4,5,6])

a * b    # output - [4, 10, 18]
```

Applications of numpy
- mathematics
- plotting
- core component in many libraries (e.g. - pandas, Digital photography)
- Machine learning (tensors)

## Data Preprocessing

### Importing the libraries

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

### Importing the dataset

````python
dataset = pd.read_csv('./data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
````

````python
print(X)
````

````python
print(y)
````

### Taking care of missing data

````python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
````

````python
print(X)
````

### Encoding categorical data

#### Encoding the Independent variable

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
````

````python
print(X)
````

#### Encoding the Dependent variable

````python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
````

````python
print(y)
````

### Splitting the dataset into the Training and Test set

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
````

````python
print(X_train)
````

````python
print(X_test)
````

````python
print(y_train)
````

````python
print(y_test)
````

### Feature Scaling

````python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
````

````python
print(X_train)
````
